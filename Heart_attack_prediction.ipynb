```python
df=pd.get_dummies(data=df)
```

```python
x=df.drop('HeartDisease',axis=1)
```

```python
y=df.HeartDisease
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
params={'n_estimators':500,"max_depth":4,"min_samples_split":5,"learning_rate":0.01,"loss":"squared_error"}
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.svm import SVC
clf=SVC(kernel='linear',C=1)
```

```python
svm_=clf.fit(x_train,y_train)
```

```python
svm_.score(x_train,y_train)
```

```python
svm_.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(10,10))
sns.distplot(y_test,color="blue",kde=True)
sns.distplot(y_pred,color="red",kde=True)
plt.title("Heart Disease")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mode11.score(x_train,y_train)
```

```python
mode11.score(x_test,y_test)
```

```python
lr_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
classification_report(y_test,lr_pred)
```

```python
displr=plot_confusion_matrix(lr,x_test,y_test,cmap='viridis',values_format='d')
```
